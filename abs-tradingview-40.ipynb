```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import lib_ga
import pickle
import os

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# Загрузка данных

```python
abs_data = pd.read_csv("../input/abscur/close_abs.csv",index_col="Date")
par_data = pd.read_csv("../input/abscur/close_pairs.csv",index_col="Date")

par_data.drop('SGDCAD',axis=1, inplace=True)

abs_data.shape,par_data.shape,abs_data.index.min(),abs_data.index.max()
```

# Генерим формулы для абсолютных курсов

```python
# функция геном в пары
def gen_to_pars(genom):
    par_index = np.argsort(list(genom.values()))[-40:]
    return np.array(list(genom.keys()))[np.sort(par_index)]
    
# цикл по абсолютам
for cur in abs_data.columns:
    print('-------------- ' + cur + ' ----------------')
    
    # выход
    y = abs_data[cur].values
    y_log = np.log(y)
    
    # кеш
    cash = {}
    
    # функция качества
    def getQualityMethod(**genom):
        # список пар
        inp_names = gen_to_pars(genom)
        
        # из кеша забираем
        # key = "-".join(list(inp_names))
        # if key in cash:
        #    return cash[key]
        
        # матрица входов
        x = par_data[inp_names].values
        x_log = np.log(x)
        
        # обучение линейной модели
        reg = LinearRegression().fit(x_log,y_log)
        
        # прогноз и экспонента
        y_pred = np.exp(reg.predict(x_log))
        
        # подсчет ошибки
        err = mean_squared_error(y,y_pred,squared=False)
        
        # в кеш кладем
        # cash[key] = -err
        
        # возвращаем отрицание ошибки
        return -err
        
    # создание модели
    modelGA = lib_ga.ListGenetic( pop_size = 200,
                                  **{par:range(100) for par in par_data.columns},
                                  quality_method=getQualityMethod)
    
    # загрузка модели, если есть
    fn = '../input/abs-tradingview-40/'+cur+'.pkl'
    if os.path.exists(fn):
        with open(fn,'rb') as fp:
            modelGA = pickle.load(fp)
    
    # поиск лучшего в генетике
    modelGA.fit(100000, echo_time=60*60, stop_time_sec=8*60*60/45)
    with open(cur+".pkl",'wb') as fp:
        pickle.dump(modelGA,fp)
    
    # ------------ лучшая модель ----------------------
    # извлекаем лучшего
    best_par = modelGA.getBestParams()
    # print(best_par)
    
    # список пар
    inp_names = gen_to_pars(best_par)
    
    # матрица входов
    x = par_data[inp_names].values
    x_log = np.log(x)
    
    # обучение линейной модели
    reg = LinearRegression().fit(x_log,y_log)
    
    # прогноз и экспонента
    y_pred = np.exp(reg.predict(x_log))
    
    # ошибка
    err = mean_squared_error(y,y_pred,squared=False)
    print("ошибка="+str(err))
    
    # извлечение коэффициентов
    # print(reg.coef_,reg.intercept_)
    
    # построение формулы
    code = 'indicator("'+cur+'ABS")\n'
    
    for n in range(len(reg.coef_)):
        code += inp_names[n] + '=request.security("'+ inp_names[n] +'",timeframe.period,close)\n'
    
    code += cur + 'ABS = math.exp(' + str(reg.intercept_)
    for n in range(len(reg.coef_)):
        if reg.coef_[n] >= 0: 
            code += '+'
        code += str(reg.coef_[n]) + '*math.log(' + inp_names[n] + ')'
    code += ')\n'
    code += 'plot('+cur+'ABS)'
    print(code)
    
    #break
```
