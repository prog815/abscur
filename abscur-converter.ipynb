# Загрузка котировок

```python
par_data=pd.read_csv("../input/abscur/close_pairs.csv",index_col="Date")
par_data
```

# Создание конвертера

```python
converter = lib_abscur.PairAbsConverter(par_data.columns.values)
converter
```

# Матрица прямого преобразования

```python
plt.figure(figsize=(20,12))
sns.heatmap(pd.DataFrame(converter.AP.astype(int),columns=par_data.columns,index=converter.Abs),cmap="PiYG",square=True,center=0,linewidths=1)
plt.title('Матрица преобразования от логарифмов абсолютных курсов к логарифмам парных курсов')
plt.show()
```

# Матрица обратного преобразования

```python
plt.figure(figsize=(12,20))
sns.heatmap(pd.DataFrame(converter.PA,index=par_data.columns.values,columns=converter.Abs),cmap="PiYG",square=True,center=0,linewidths=1)
plt.title('Матрица преобразования от логарифмов парных курсов к логарифмам абсолютных курсов')
plt.show()
```

# Проверка произведения прямой и псевдообратной матриц

```python
E = np.matmul(converter.AP,converter.PA)
plt.figure(figsize=(15,15))
sns.heatmap(E,cmap="PiYG",square=True,center=0,xticklabels=converter.Abs,yticklabels=converter.Abs,linewidths=1)
plt.title('Матрица произведения прямой и псевдообратной')
plt.show()
```

```python
np.unique(np.round(E.reshape(-1),12),return_counts=True)
```

```python
plt.figure(figsize=(12,5))
_=plt.hist(E.reshape(-1),bins=100)
plt.grid()
plt.show()
```
