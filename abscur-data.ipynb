```python
import pandas as pd
import requests
import tempfile
```

# Абсолютные валютные курсы

Импортируем данные с [**листа Google-таблицы**](https://docs.google.com/spreadsheets/d/1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8/edit#gid=1871847763). Данные на листе обновляются каждый день.

```python
url="https://docs.google.com/spreadsheets/d/1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8/export?format=csv&id=1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8&gid=1871847763"
text = requests.get(url).text
with tempfile.NamedTemporaryFile("wt",delete=False) as tmp:
    tmp.write(text)
    tmp.close()
    data_abs = pd.read_csv(tmp.name,index_col="Date",decimal=",",parse_dates=True)
    
data_abs.to_csv("abscur.csv")
data_abs
```

# Расшифровка валют

Импортируем данные с листа [**Google-таблицы**](https://docs.google.com/spreadsheets/d/1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8/edit#gid=1419524366).

```python
url="https://docs.google.com/spreadsheets/d/1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8/export?format=csv&id=1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8&gid=1419524366"
cur_ras = pd.read_csv(url,index_col=0).iloc[:,:-1]
cur_ras.to_csv("cur_ras.csv")

def make_clickable(val):
    # target _blank to open new window
    return '<a target="_blank" href="{}">Wikipedia link</a>'.format(val)

cur_ras.style.format({'Ссылка на Wikipedia': make_clickable})
```
