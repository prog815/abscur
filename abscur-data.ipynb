Страница производит расчет абсолютных валютных курсов на основе парных валютных курсов. Мы используем математическую оптимизацию и систему уравнений из логарифмов парных курсов и абсолютных валютных курсов для рассчета абсолютных курсов на ежедневной основе. 

Расчет абсолютных валютных курсов в каждом дне производим по следующей методике. Из предположения что любой парный валютный курс представлякется как отношение двух абсолютных валютных курсов

$$
P_n = \frac{A_i}{A_j}
$$

применяя логарифмирование получаем систему линейных уравнений

$$
\begin{equation}
\begin{aligned}
    ln(P_1) &= ln(A_1)-ln(A_2) \\
    ...\\
    ln(P_n) &= ln(A_i)-ln(A_j) \\
    ...\\
\end{aligned}
\end{equation}
$$
, где $P_n$ - парный валютных курс для некоторой пары валют под номером $n$, $A_i$ - абсолютный валютный курс для некотрой валюты под номером $i$.
Левая часть уравнений с логарифмами парных курсов нам известна. Остается найти правую часть с логарифмами абсолютных курсов. Находим их методом наименьших квадратов. Затем применив экспоненту получаем значения для абсолютных курсов.

$$
A_i = e^{ln(A_i)}
$$

# Абсолютные валютные курсы

Не импортируем данные с [**листа Google-таблицы**](https://docs.google.com/spreadsheets/d/1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8/edit#gid=1871847763) как раньше (данные на листе обновляются каждый день).

Абсолютные курсы считаем из парных валютных курсов аппроксимацией. Переход к новой методике расчета проанализирован в тетрадке https://www.kaggle.com/code/eavprog/abscur-approks.

Методика расчета следующая:
1. Считаем 

```python
# Старая методика

# url="https://docs.google.com/spreadsheets/d/1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8/export?format=csv&id=1UgC2NfKtKJPaXglsmVJET7Xnp2ZtOzJgKD9DN-VGFA8&gid=1871847763"
# text = requests.get(url).text
# with tempfile.NamedTemporaryFile("wt",delete=False) as tmp:
#     tmp.write(text)
#     tmp.close()
#     data_abs = pd.read_csv(tmp.name,index_col="Date",decimal=",",parse_dates=True)
    
# data_abs.to_csv("abscur.csv")
# data_abs

# ---------------------------

# Новая методика

# Конвертируем имена пар в вид со знаком дроби
par_names = [par[:3] + "/" + par[3:] for par in data_par.columns]
# print(par_names)

# Извлекаем имена валют из имен валютных пар

cur_names = []
for par in par_names:
    cur1,cur2 = par.split('/')
    if cur1 not in cur_names:
        cur_names.append(cur1)
    if cur2 not in cur_names:
        cur_names.append(cur2)
# print(cur_names)

import numpy as np

# Строим матрицу перехода от абсолютныых курсов к парным
M = np.zeros([len(par_names),len(cur_names)])
for par in par_names:
    cur1,cur2 = par.split('/')
    par_index = par_names.index(par)
    cur1_index = cur_names.index(cur1)
    cur2_index = cur_names.index(cur2)
    M[par_index,cur1_index] = 1
    M[par_index,cur2_index] = -1
# M 

# Считаем абсолютные курсы аппроксимацией
abscur = pd.DataFrame(columns=cur_names,index=data_par.index)
for i in range(data_par.shape[0]):
    lnA, _, _, _ = np.linalg.lstsq(M, np.log(data_par.iloc[i,:].values), rcond=None)
    abscur.iloc[i,:] = np.exp(lnA)

abscur.to_csv("abscur.csv")
abscur
```
