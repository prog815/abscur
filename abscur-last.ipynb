```python
import os

def find(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)
```

# История парных курсов

```python
close_par = pd.read_pickle(find('close_pairs.pkl','..'))
close_par
```

```python
# _ = lib_abscur_post.sendTable(close_par,"par_hist")
```

# Последние парные курсы

```python
d = pd.DataFrame(index=pd.date_range(close_par.index[0],close_par.index[-1]),columns=close_par.columns)
d.loc[close_par.index.values] = close_par.values
d.fillna(method='pad',inplace=True)

df = pd.DataFrame()
df[f'курс на {close_par.index[-1]}'] = d.iloc[-1,:]
df['% изм.день'] = d.pct_change(1).iloc[-1,:]
df['% изм.неделя'] = d.pct_change(7).iloc[-1,:]
df['% изм.месяц'] = d.pct_change(30).iloc[-1,:]
df['% изм.квартал'] = d.pct_change(90).iloc[-1,:]
df['% изм.полгода'] = d.pct_change(180).iloc[-1,:]
df['% изм.год'] = d.pct_change(365).iloc[-1,:]
df
```

```python
df.to_csv("par_last.csv")
```
