Данные для настоящего расчета берутся из тетрадки [**Источник данных для проекта "Абсолютные валютные курсы"**](https://www.kaggle.com/code/eavprog/abscur-data).

В настоящем расчете показываем абсолютные курсы на последнюю имеющуюся дату. Показываем расшифровку тикера валюты и ссылку на страницу с описанием на Wikipedia. Курсы обновляются ежедневно.

```python
import abscur_menu
abscur_menu.print_menu()
```

```python
import pandas as pd
import datetime

abscur = pd.read_csv('/kaggle/input/abscur-data/abscur.csv',index_col=0,parse_dates=True)
last_abscur = abscur.last("1D").transpose()

last_abscur = last_abscur.sort_index()

last_abscur.columns = ["курс на " + datetime.datetime.strftime(last_abscur.columns[0],"%Y-%m-%d")]

abscur_ras = pd.read_csv('/kaggle/input/abscur-data/cur_ras.csv',index_col=0)

def make_clickable(val):
        return '<a target="_blank" href="{}">Wikipedia link</a>'.format(val)


    
last_abscur.merge(abscur_ras,left_index=True,right_index=True).style.format({'Ссылка на Wikipedia': make_clickable})
```

```python
import abscur_post as ap

res = last_abscur.merge(abscur_ras,left_index=True,right_index=True)
text = f"Абсолютные курсы на {res.columns[0][-10:]}:\n"
for n in range(len(res)):
    ln = res.iloc[n]
    text += f"{res.index[n]} - {ln[0]} - {ln[1]}\n"
text += f"---\nИсточник - https://www.kaggle.com/code/eavprog/abscur-posled-kurs"
# print(text)
ap.send_text(text)
```
