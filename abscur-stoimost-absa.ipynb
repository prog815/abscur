# Стоимость Абса

В настоящем проекте принято выражать валюты, акции и другие инструменты по абсолютному курсу, т.е. в Абсах. А сколько стоит один Абс через другие валюты? Для вычисления возьмем и обратим дробь с абсолютным валютным курсом (см. https://www.kaggle.com/eavprog/abscur-posled-kurs).

Например, если мы хотим узнать стоимость одного Абса в долларах, то возьмем курс доллара к Абсу и обратим его: 1 / курс доллара к Абсу. Таким образом, если курс доллара к Абсу равен 20.72, то стоимость одного Абса в долларах будет 0.048 доллара.

Аналогично можно вычислить стоимость Абса в других валютах или инструментах. Например, для вычисления стоимости Абса в евро нужно взять обратный курс евро к Абсу.

Важно учитывать, что курсы валют постоянно меняются и могут быть различными на разных биржах и площадках. При вычислениях стоимости Абса следует использовать актуальные данные о курсах валют и других инструментов.

```python
import numpy as np
import pandas as pd
import datetime
import abscur_menu
import abscur_post as ap
from IPython.display import Markdown, display

# меню проекта печатаем
abscur_menu.print_menu()

# загрузка истории абсолютных курсов
abscur = pd.read_csv("/kaggle/input/abscur-data/abscur.csv",index_col=0,parse_dates=['Date'])

# загрузка расшифровки валют
ras_cur = pd.read_csv("/kaggle/input/abscur-data/cur_ras.csv",index_col=0)
```

```python
names = list(abscur.columns)
names.sort()
text = "# Стоимость Абса на " + datetime.datetime.strftime(abscur.index[-1],"%Y-%m-%d") + "\n\n"
text_post = "Стоимость Абса на " + datetime.datetime.strftime(abscur.index[-1],"%Y-%m-%d") + "\n---\n"
for name in names:
    text += "- 1 **Абс** = " + str(np.round(1/abscur[name][-1],3)) + " **" + name + "** ([**" + ras_cur.loc[name][0] + "**](" + ras_cur.loc[name][1]+ "))" + " (см. [график истории абсолютного курса " + name + "](https://www.kaggle.com/code/eavprog/abscur-hist/notebook#" + name + "))\n"
    text_post += "1 Абс = " + str(np.round(1/abscur[name][-1],3)) + " " + name + " (" + ras_cur.loc[name][0] + ")" + "\n"
display(Markdown(text))
text_post += "---\nИсточник - https://www.kaggle.com/code/eavprog/abscur-stoimost-absa"
ap.send_text(text_post)
```
