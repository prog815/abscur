# Стоимость Абса

В настоящем проекте принято выражать валюты, акции и другие инструменты по абсолютному курсу, т.е. в Абсах. А сколько стоит один Абс через другие валюты? Для вычисления возьмем обратим дробь.

```python
import numpy as np
import pandas as pd
import datetime
import abscur_menu
from IPython.display import Markdown, display

# меню проекта печатаем
abscur_menu.print_menu()

# загрузка истории абсолютных курсов
abscur = pd.read_csv("/kaggle/input/abscur-data/abscur.csv",index_col=0,parse_dates=['Date'])

# загрузка расшифровки валют
ras_cur = pd.read_csv("/kaggle/input/abscur-data/cur_ras.csv",index_col=0)
```

```python
names = list(abscur.columns)
names.sort()
text = "# Стоимость Абса на " + datetime.datetime.strftime(abscur.index[-1],"%Y-%m-%d") + "\n\n"
for name in names:
    text += "1 **Абс** = " + str(np.round(1/abscur[name][-1],3)) + " **" + name + "** ([" + ras_cur.loc[name][0] + "](" + ras_cur.loc[name][1]+ "))" + "\n\n"

display(Markdown(text))
```
