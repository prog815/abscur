```python
import numpy as np
import pandas as pd
import yfinance as yf
import lib_abscur
```

# Список валютных пар

```python
curList = 'AUDCAD	AUDHKD	AUDJPY	AUDSEK	AUDUSD	CADJPY	CADSEK	CHFAUD	CHFHKD	CHFJPY	EURAUD	EURCAD	EURCHF	EURCNY	EURCZK	EURGBP	EURHKD	EURILS	EURJPY	EURNOK	EURNZD	EURRUB	EURSEK	EURSGD	EURUSD	EURZAR	GBPAUD	GBPCAD	GBPCHF	GBPHKD	GBPJPY	GBPNOK	GBPSEK	GBPSGD	GBPUSD	HKDSEK	HKDSGD	JPYHKD	JPYSEK	NZDSEK	NZDUSD	SGDAUD	SGDCAD	SGDCHF	SGDJPY	USDAED	USDARS	USDBRL	USDCAD	USDCHF	USDCLP	USDCNY	USDCOP	USDCZK	USDDKK	USDEGP	USDHKD	USDHUF	USDIDR	USDILS	USDINR	USDISK	USDJPY	USDKRW	USDKWD	USDKZT	USDMXN	USDMYR	USDNOK	USDPEN	USDPHP	USDPKR	USDPLN	USDQAR	USDRON	USDRUB	USDSAR	USDSEK	USDSGD	USDTHB	USDTRY	USDTWD	USDUAH	USDVND	USDZAR	USDGBP	GBPRUB'.split('\t')
print(curList)
```

# Предыдущие значения

```python
import os

def find(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)
        
prev_close_pairs_file_path = find('close_pairs.pkl','..')
prev_close_pairs_file_path
```

```python
# last_closePairs = pd.read_csv('../input/abscur/close_pairs.csv',index_col='Date',header=0,decimal='.')
last_closePairs = pd.read_pickle(prev_close_pairs_file_path)
last_closePairs.index = last_closePairs.index.astype(str)
last_closePairs
```

```python
from datetime import datetime, timedelta
yesterday = datetime.now() - timedelta(1)
yesterday = datetime.strftime(yesterday, '%Y-%m-%d')
yesterday
```

# Загрузка парных валютных курсов

```python
try:
    all_data = yf.download([c + '=X' for c in curList],period='5y',threads=True,end=yesterday)
    all_data.index = pd.to_datetime(all_data.index,format='%Y-%m-%d').astype(str)
    all_data = all_data[all_data.index <= yesterday]
except:
    print("error")
    all_data = None
```

```python
all_data
```

# Выделяем курсы закрытия для валютных пар

```python
if(all_data is None):
    closePairs = last_closePairs
else:
    closePairs = all_data.Close
    closePairs.columns = [c.split('=')[0] for c in closePairs.columns]
    closePairs
```

# Заполняем пустоты из предыдущей версии

```python
for cur in curList:
    c = closePairs[cur]
    lc = last_closePairs[cur]
    
    c_index = c.index[c.isna()]
    ints_index = lc.index.intersection(c_index)
#     if(len(ints_index)>0):
    c[ints_index] = lc[ints_index].values
        
closePairs
```

# Выгружаем курсы закрытия на диск (выход тетрадки)

```python
closePairs.fillna(method='pad',axis=0,inplace=True)
closePairs.to_csv('close_pairs.csv')
closePairs.to_pickle('close_pairs.pkl')
closePairs
```

# Расчет абсолютных курсов

```python
converter = lib_abscur.PairAbsConverter(list(closePairs.columns))
closeAbs = pd.DataFrame(converter.pairs2abs(closePairs.values),columns=converter.Abs,index=closePairs.index)
closeAbs.to_csv('close_abs.csv')
closeAbs.to_pickle('close_abs.pkl')
closeAbs
```
