```
from abscur.optim import ListGenetic
```

# определяем данные

```
import numpy as np

N = 1000

X = np.random.randn(N)
Y = np.random.randn(N)

A = 2
B = 4
C = 6

Z = A*X+B*Y+C

Z += np.random.randn(N)*0.1
```

# запускаем генетический поиск

```
def getQualityMethod(a,b,c):
    z = a*X+b*Y+c
    return -np.std(Z-z)

modelGA = ListGenetic(pop_size=20,
                        a=np.arange(-10,10),
                        b=np.arange(-10,10),
                        c=np.arange(-10,10),
                        quality_method=getQualityMethod)

modelGA.fit(500)
```

# выводим результаты

```
modelGA.plot_hist_new(('quality','a','b','c'))
```
